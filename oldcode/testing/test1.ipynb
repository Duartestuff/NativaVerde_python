```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


df = pd.read_csv('../database/Siembras.csv')

df['Fecha'] = pd.to_datetime(df['Fecha'], dayfirst=True)

dfMedellin = df.loc[
    (df['Ciudad'] == 'Medellín') & (df['Arboles'] > 200)
].sort_values(by='Fecha')

dfMedellin
```

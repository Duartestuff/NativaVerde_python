```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../database/Siembras.csv')

dfSantaRosaOsos = df.loc[
    (df['Ciudad'] == 'Santa Rosa de Osos') & ((df['Vereda'] == 'Las Cruces') | (df['Vereda'] == 'Mina Vieja'))
]

data = {
    "Vereda": dfSantaRosaOsos['Vereda'],
    "Arboles": dfSantaRosaOsos['Arboles']
}

plt.plot(data['Vereda'], data['Arboles'])
plt.title('Información')
plt.grid(True)
plt.show()
```

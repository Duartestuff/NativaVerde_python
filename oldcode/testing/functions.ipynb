```python
import pandas as pd

def filtroPorCiudadVereda(df, ciudad, vereda):

    if(type(vereda) == list):
        finalDF = df.loc[
                    (df['Ciudad'] == ciudad) & (df['Vereda'].isin(vereda))
                ]
        return finalDF
    
    else:
        finalDF = df.loc[
                    (df['Ciudad'] == ciudad) & (df['Vereda'] == vereda)
                ]
        return finalDF

def filtroPorCiudadNumero(df, ciudad, numero):
    
    return df.loc[
        (df['Ciudad'] == ciudad) & (df['Arboles'] > numero)
    ]
```

Para el filtro de veredas:

```python
df = pd.read_csv('../database/Siembras.csv')
newdf = filtroPorCiudadVereda(df, 'Medellín', ['Aguas Frias', 'Boqueron'])

newdf
```

Para el fitro de numero de árboles por ciudad

```python
numberDf = filtroPorCiudadNumero(df, 'Medellín', 1000)
numberDf
```

```python
import pandas as pd
import matplotlib.pyplot as plt

filtro_graficas = ['Fecha', 'Arboles']
orden_columnas_data = ['Vereda', 'Arboles', 'Hectareas', 'Fecha', 'Nombre comun', 'evento']
data = pd.read_csv('../database/Siembras.csv')

data_filtro = data.loc[(data['Ciudad'] == 'Medellín')]

data_filtro[orden_columnas_data]
```

```python
data_grafica = data_filtro[filtro_graficas]

data_grafica['Fecha'] = pd.to_datetime(data_grafica['Fecha'])

aggregated_data = data_grafica.groupby('Fecha')['Arboles'].sum()

aggregated_data
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(aggregated_data.index, aggregated_data.values, marker='o')
plt.plot(aggregated_data.index, aggregated_data.values, color='skyblue', alpha=0.7)

plt.xlabel('Fecha')
plt.ylabel('Árboles sembrados')
plt.title('Árboles sembrados en')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
